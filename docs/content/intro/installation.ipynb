# Встановлення і робота з інтерпретатором Python

- Зайти на сайт [Python](http://python.org/downloads/) і завантажити свіжу версію інтерпретатора, яка співвідноситься з ОС, встановленої на комп'ютері.

![](..\images\intro\Python_install.bmp)

- Запустити інтегровану середу Python IDLE.

![](..\images\intro\IDLE_Open.bmp)
     
- З'явиться вікно середовища інтерпретатора, готового до інтерактивної роботи.

![](..\images\intro\IDLE_Opened.bmp)

Середовище IDLE забезпечує три ключових можливості: 
* Введення виразів і програмного коду на мові Python з отриманням результатів прямо в командній оболонці Python; 
* Надає редактор програмного коду з підсвічуванням синтаксису мови Python і підтримкою функції оформлення відступів 
* Покрокове виконання програмного коду яке полегшує пошук і усунення помилок. 
<!--М. замінити на опис можливостей, які більш можливі за потужний калькулятор-->
Інтегроване середовище розробки особливо зручне при опробуванні простих алгоритмів, фрагментів програмного коду і регулярних виразів і може використовуватися як дуже потужний і гнучкий калькулятор.

Програмний код на мові Python можна записати за допомогою будь- якого простого текстового редактора, який здатний завантажувати і зберігати текст або в кодуванні ASCII, або UTF-8. 
За замовчуванням передбачається, що файли на мові Python зберігаються в кодуванні UTF-8, підмножині кодування ASCII, за допомогою якої можна представити практично будь-який символ алфавіту.

Файл на мові Python можна запустити наступним чином:

- Створити файл в IDLE File->New File

    Приклад:

![](..\images\intro\IDLE_Code.bmp)

- Для запуску натиснути Run->Run Module :

    Якщо є помилки, вони будуть локалізовані.

![](..\images\intro\IDLE_Error.bmp)

Файл або код також можна протестувати за допомогою терміналу. 

- Відкрийте термінал

    *Windows: Win+R, введіть cmd так натисніть ok*

    *Linux: Ctrl+Alt+T*

- Введіть python (у разі помилки: https://geek-university.com/python/add-python-to-the-windows-path/)
- Построково скопіюйте та вставте свій код

Окрім терміналу та IDLE також існують спеціальні редактори коду: Notepad++, SumblimeText, VisualCode, та середовище розробки PyCharm [Докладніше](http://link)
